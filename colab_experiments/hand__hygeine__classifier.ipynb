```python
from google.colab import drive
drive.mount('/content/drive')

%cd /content/drive/MyDrive/hand_hygeine_project
!pip install -U ultralytics

yaml_text = """
path: /content/drive/MyDrive/hand_hygeine_project/dataset
train: images/train
val: images/train

names:
  0: Gel
  1: Hand
"""

with open("/content/drive/MyDrive/hand_hygeine_project/data.yaml", "w") as f:
    f.write(yaml_text)

from ultralytics import YOLO

model = YOLO("ultralytics/cfg/models/v11/yolo11-seg.yaml")

model.train(
    data="/content/drive/MyDrive/hand_hygeine_project/data.yaml",
    epochs=50,
    imgsz=640,
    batch=8,
    project="/content/drive/MyDrive/hand_hygeine_project/runs",
    name="hand_hygiene_seg",
    device=0
)
```
