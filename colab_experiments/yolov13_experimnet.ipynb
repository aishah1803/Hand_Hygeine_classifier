```python
import os
from pathlib import Path
from google.colab import drive

# Connect to Google Drive
drive.mount('/content/drive')

# Tell Python where your dataset folder is
DATASET_DIR = "/content/drive/MyDrive/label studio"

# Go into that folder
os.chdir(DATASET_DIR)

# Confirm it worked
print("We are now in:", os.getcwd())
```

```python
# Create the settings file that tells YOLOv13 where your data is

yaml_text = f"""
path: {DATASET_DIR}
train: images/train
val: images/validate
test: images/test

nc: 2
names: ['Hand', 'Gel']
"""

with open("hand_segmentation_dataset.yaml", "w") as f:
    f.write(yaml_text)

yaml_path = str(Path("hand_segmentation_dataset.yaml").resolve())

print("YAML file created successfully")
print("Location:", yaml_path)
print("Contents:", yaml_text)
```

```python
from ultralytics import YOLO

# Load the YOLOv13 nano model
# Colab will automatically download it for you
model = YOLO("yolov13n.pt")

print("YOLOv13 model loaded successfully")
```

```python
!pip install -q "sympy==1.13.3"
```

```python
import torch

results = model.train(
    task="segment",
    data=yaml_path,
    epochs=20,
    imgsz=640,
    batch=2,
    device="0" if torch.cuda.is_available() else "cpu",
    project="hand_seg_runs_v13",
    name="yolov13n_seg"
)

print("Training finished!")
```

```python
from pathlib import Path
from google.colab import files

# Find the best weights file
best_weights = Path("hand_seg_runs_v13/yolov13n_seg2/weights/best.pt")

print("File exists?", best_weights.exists())
print("File size:", round(best_weights.stat().st_size / 1024 / 1024, 2), "MB")

# Download it to your laptop
files.download(str(best_weights))
```

```python
from ultralytics import YOLO
from pathlib import Path

# Load your trained YOLOv13 model
model = YOLO("hand_seg_runs_v13/yolov13n_seg2/weights/best.pt")

# Run prediction on one of your test images
# Change this path to any image in your test folder
test_image = "/content/drive/MyDrive/label studio/images/test"

results = model.predict(
    source=test_image,   # folder of test images
    imgsz=640,
    conf=0.3,            # confidence threshold
    save=True,           # save output images with boxes drawn
    project="test_output_v13",
    name="predictions"
)

print("Prediction done!")
print("Number of images processed:", len(results))

# Show what was detected in each image
for i, result in enumerate(results):
    print(f"\nImage {i+1}:")
    if result.boxes is not None and len(result.boxes) > 0:
        for box in result.boxes:
            class_id = int(box.cls)
            confidence = float(box.conf)
            class_name = result.names[class_id]
            print(f"  Detected: {class_name} with confidence {confidence:.2f}")
    else:
        print("  Nothing detected")
```
